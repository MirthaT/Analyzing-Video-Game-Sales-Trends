## Tratar los valores ausentes

Las columnas 'model_year', 'cylinders', 'odometer', y 'is_4wd' son numéricas por lo que rellenar los valosres ausentes utilizando la mediana podría ser la razon mas razonable ya que es menos sensibles a valores atípicos que el promedio.

Columna 'paint_color'
Dado que la columna 'paint_color' es categórica, dejaré los valores ausentes como están, ya que no es crucial para todas las operaciones de análisis que se realizará.

```python
# Tratando valores ausentes en 'model_year','cylinders' y 'odometer'
median_model_year = vehicles['model_year'].median()
median_cylinders = vehicles['cylinders'].median()
median_odometer = vehicles['odometer'].median()
vehicles['model_year'].fillna(median_model_year, inplace=True)
vehicles['cylinders'].fillna(median_cylinders, inplace=True)
vehicles['odometer'].fillna(median_odometer, inplace=True)
```

## Enriquecer datos

Para enriquecer los datos con más información temporal, se extraerá el día de la semana, el mes y el año en que se colocó el anuncio a partir de la columna 'date_posted'

```python
# Agrega valores de fecha y hora para cuando se colocó el anuncio

vehicles['day_of_week'] = vehicles['date_posted'].dt.dayofweek
vehicles['month'] = vehicles['date_posted'].dt.month
vehicles['year'] = vehicles['date_posted'].dt.year
```

```python
# Agrega los años del vehículo cuando el anuncio se colocó
vehicles['vehicle_date_posted'] = vehicles['year'] - vehicles['model_year']
```

```python
# Agrega el millaje promedio del vehículo por año

# Calcula el valor promedio de millaje por año
millaje_year = vehicles.groupby('model_year')['odometer'].mean().reset_index()
millaje_year.rename(columns={'odometer': 'avg_millaje_year'}, inplace=True)

# Combina los datos originales con los valores de millaje promedio por año
vehicles = vehicles.merge(millaje_year, on='model_year', how='left')
```

```python
#veamos los valores unicos de la columna 'condition'
vehicles["condition"].unique()
```

```python
# Podría ser útil reemplazar los valores de condición con algo que se pueda manipular más fácilmente
#Reemplacemos los valores de la columna 'condition' con valores numéricos en una escala de 0 a 5

def condition_to_scale(valor):
    if valor == "excellent":
        return 5
    elif valor == "good":
        return 4
    elif valor == "like new":
        return 3
    elif valor == "fair":
        return 2
    elif valor == "new":
        return 1
    elif valor == "salvage":
        return 0

# Aplicar la función lambda a la columna 'condition'
vehicles['condition'] = vehicles['condition'].apply(lambda x: condition_to_scale(x))
```

## Estudiar y tratar valores atípicos

Determinando los límites superiores e inferiores de los valores atípicos, y creando un Dataframe filtrando los registros atipicos. 

```python
# Determina los límites inferiores para valores atípicos
lower_limits = vehicles[['price', 'odometer', 'model_year']].apply(lambda x: x.quantile(0.01))
```

```python
# Determinar los límites superiores para valores atípicos
upper_limits = vehicles[['price', 'odometer', 'model_year']].apply(lambda x: x.quantile(0.95))
```

```python
# Filtrar los datos sin valores atípicos en un DataFrame separado
filtered_vehicles = vehicles[
    (vehicles['price'] >= lower_limits['price']) &
    (vehicles['price'] <= upper_limits['price']) &
    (vehicles['odometer'] >= lower_limits['odometer']) &
    (vehicles['odometer'] <= upper_limits['odometer']) &
    (vehicles['model_year'] >= lower_limits['model_year']) &
    (vehicles['model_year'] <= upper_limits['model_year'])
]
```

```python
#Verificando el tamaño del Dataframe original y el filtrado
vehicles.shape[0],filtered_vehicles.shape[0]
```

## Estudiar parámetros principales sin valores atípicos

Hagamos un análisis  para visualizar y comparar la distribución de las columnas numéricas con y sin valores atípicos, lo que puede proporcionar información valiosa sobre cómo los valores atípicos pueden afectar las estadísticas y la interpretación de los datos.

```python
for col in numerical_columns:
  fig, ax = plt.subplots(ncols=2, figsize=((10, 4)))


  outlier_threshold = vehicles[col].quantile(q=0.95)

  ax[0].set_title(f"Histogram of {col} (all observations)")
  vehicles[col].hist(bins=50, ax=ax[0])
  ax[0].set_xlabel(col)

  ax[1].set_title(f"Histogram of {col} (without outliers)")
  vehicles[(vehicles[col] < outlier_threshold) & (vehicles[col] > 0)][col].hist(bins=50, ax=ax[1])
  ax[1].set_xlabel(col)
  fig.tight_layout(pad=2.0)
```

Histograma "without outliers": En este histograma, donde he filtrado los valores atípicos según el umbral del percentil 95 y el límite inferior, vemos que la distribución es más compacta y concentrada en un rango más estrecho de precios, principalmente entre 0 y 30000. La barra con frecuencia más alta se encuentra en el rango de 2300, lo que indica que hay un grupo significativo de anuncios con precios en esa área.

En resumen, las conclusiones que puedos obtener de estos histogramas son que la mayoría de los anuncios tienen precios relativamente bajos y se distribuyen principalmente en el rango de 0 a 30000. ## Periodo de colocación de los anuncios

identificar anuncios que se eliminaron rápidamente o estuvieron publicados por un tiempo anormalmente largo utilizando umbrales basados en la media y la desviación estándar:

```python
#Calulando la media, la mediana y la desviación estándar
filtered_vehicles['days_listed'].describe()[['mean','50%','std']]
```

```python
# Calcular el número de anuncios y el precio promedio para cada modelo de vehículo
vehicle_counts = filtered_vehicles['model'].value_counts()
vehicle_prices = filtered_vehicles.groupby('model')['price'].mean()

# Seleccionar los dos modelos con el mayor número de anuncios
top_models = vehicle_counts.head(2).index

# Crear un gráfico de barras para mostrar la cantidad de anuncios en cada modelo de vehículo
plt.figure(figsize=(8, 4))
sns.countplot(data=filtered_vehicles, y='model', order=top_models)
plt.title('Número de Anuncios por Modelo de Vehículo')
plt.xlabel('Número de Anuncios')
plt.ylabel('Modelo de Vehículo')
plt.show()

# Imprimir el número de anuncios y el precio promedio para los dos modelos principales
for model in top_models:
    print(f'Modelo de Vehículo: {model}')
    print(f'Número de Anuncios: {vehicle_counts.loc[model]}')
    print(f'Precio Promedio: ${vehicle_prices.loc[model]:.2f}\n')
```

## Factores de precio

Para analizar los factores que impactan en el precio, puedes seguir estos pasos:

Seleccionaré los dos tipos más populares de vehículos que detectaste anteriormente.
Filtraré los datos para incluir solo los anuncios de los tipos seleccionados.
Analizaré cómo el precio depende de la edad, millaje, condición, tipo de transmisión y color.
Crearé gráficos de dispersión para las variables numéricas y gráficos de caja y bigotes para las variables categóricas.
Utilizaré una matriz de correlación y diagramas de correlación para analizar las relaciones entre las variables.

```python
# Filtrar los datos para los tipos de vehículos populares
popular_vehicles = filtered_vehicles[filtered_vehicles['model'].isin(top_models)]

# Variables categóricas para análisis
categorical_columns = ['transmission', 'paint_color']

# Variables numéricas para análisis
numeric_columns = ['model_year', 'odometer', 'condition','price']

# Crear gráficos de caja y bigotes para las variables categóricas con al menos 50 anuncios
for col in categorical_columns:
    value_counts = popular_vehicles[col].value_counts()
    filtered_ads = value_counts[value_counts >= 50].index
    if len(filtered_ads) > 1:
        plt.figure(figsize=(10, 6))
        popular_vehicles[popular_vehicles[col].isin(filtered_ads)].boxplot(column='price', by=col)
        plt.title(f'Boxplot de Precio por {col}')
        plt.xlabel(col)
        plt.ylabel('Precio')
        plt.xticks(rotation=45)
        plt.show()
# Crear gráficos de dispersión hexbin para las variables numéricas
for col in numeric_columns:
    plt.figure(figsize=(10, 6))
    plt.hexbin(popular_vehicles[col], popular_vehicles['price'], gridsize=20, cmap = 'Blues')
    plt.title(f'Gráfico de Dispersión Hexbin de Precio vs {col}')
    plt.xlabel(col)
    plt.ylabel('Precio')
    plt.colorbar(label='Densidad de Puntos')
    plt.show()

# Calcular la matriz de correlación para las variables numéricas
numeric_data = popular_vehicles[numeric_columns]
correlation_matrix = numeric_data.corr()

# Imprimir la matriz de correlación
print("Matriz de Correlación:")
print(correlation_matrix)
```
