# Descripción del Proyecto


La cadena de gimnasios Model Fitness está desarrollando una estrategia de interacción con clientes basada en datos analíticos.
Uno de los problemas más comunes que enfrentan los gimnasios y otros servicios es la pérdida de clientes. ¿Cómo descubres si un/a cliente ya no está contigo? Puedes calcular la pérdida en función de las personas que se deshacen de sus cuentas o no renuevan sus contratos. Sin embargo, a veces no es obvio que un/a cliente se haya ido: puede que se vaya de puntillas.

Los indicadores de pérdida varían de un campo a otro. Si un usuario o una usuaria compra en una tienda en línea con poca frecuencia, pero con regularidad, no se puede decir que ha huido. Pero si durante dos semanas no ha abierto un canal que se actualiza a diario, es motivo de preocupación: es posible que tu seguidor o seguidor/a se haya aburrido y te haya abandonado.

En el caso de un gimnasio, tiene sentido decir que un/a cliente se ha ido si no viene durante un mes. Por supuesto, es posible que estén en Cancún y retomen sus visitas cuando regresen, pero ese no es un caso típico. Por lo general, si un/a cliente se une, viene varias veces y luego desaparece, es poco probable que regrese.
Con el fin de combatir la cancelación, Model Fitness ha digitalizado varios de sus perfiles de clientes. Tu tarea consiste en analizarlos y elaborar una estrategia de retención de clientes.
Tienes que:

  * Aprender a predecir la probabilidad de pérdida (para el próximo mes) para cada cliente.
  * Elaborar retratos de usuarios típicos: selecciona los grupos más destacados y describe sus características      principales.
  * Analizar los factores que más impactan la pérdida.
  * Sacar conclusiones básicas y elaborar recomendaciones para mejorar la atención al cliente:
    identificar a los grupos objetivo;
    sugerir medidas para reducir la rotación;
    describir cualquier otro patrón que observes con respecto a la interacción con los clientes.

## Construir un modelo para predecir la cancelación de usuarios

### Crea un modelo de clasificación binaria para clientes donde la característica objetivo 

```python
# divide los datos en características (la matriz X) y una variable objetivo (y)
X = gym.drop(columns = ['Churn'])
y = gym['Churn']
```

```python
# divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=200)
```

### Entrena el modelo en el grupo de entrenamiento con dos métodos:
1. regresión logística;
2. bosque aleatorio.

```python
#Estandarizacion de datos
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

lr_model = LogisticRegression(max_iter=100, random_state=200)
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=200)
```

```python
# Entrenar
lr_model.fit(X_train_scaled, y_train)
```

```python
rf_model.fit(X_train, y_train)
```

```python
## Prediccion de probabilidad
```

```python
lr_preds = lr_model.predict(X_test_scaled)
rf_preds = rf_model.predict(X_test)
rf_preds
```

## Crear clústeres de usuarios/as

### Aplicar el dendrograma para visualizar la estructura jerárquica de las caracteristicas

```python
linked = linkage(X_train_scaled, method = 'ward') 
```

```python
# Crear el dendrograma
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top', no_labels=True)
plt.title('Agrupación jerárquica para Gym')
plt.show()
```

### Ajustar un modelo de K-Means con 5 clústeres en tus datos de entrenamiento.

```python
# define el modelo k_means con 5 clústeres
km = KMeans(n_clusters=5, random_state=200,n_init=10)
km.fit(X_train_scaled)
```

```python
km.cluster_centers_
```

### Invertir la escala de los centroides

Al transformar los centroides nuevamente a la escala original, se interpretará mejor las características en términos de sus valores originales


```python
scaled_centroids_gym = pd.DataFrame(km.cluster_centers_, columns=X_train.columns)
scaled_centroids_gym['centroid'] = scaled_centroids_gym.index

centroids_df = pd.DataFrame(scaler.inverse_transform(km.cluster_centers_), columns=X_train.columns)
centroids_df['centroid'] = centroids_df.index
centroids_df
```

2. Near_Location:
Clúster 1 está muy cerca de la ubicación.
Clúster 3 está ligeramente por debajo del promedio en este aspecto.

3. Partner y Promo_friends:
Clúster 2 tiene una alta proporción de socios y personas que han recibido descuentos por recomendaciones de amigos.
Clúster 1 tiene una baja proporción en ambos.

4. Contract_period, y Month_to_end_contract:
Clúster 0 tiene un período de contrato y meses hasta el fin del contrato relativamente altos.
Clúster 3 tiene un período de contrato y meses hasta el fin del contrato más bajos.

5. Lifetime y Age:
Clúster 2 tiene una duración de membresía más alta y una edad más alta.
Clúster 1 tiene una duración de membresía más baja y una edad más baja.

6. Avg_class_frequency_total y	Avg_class_frequency_current_month:
Clúster 0 tiene frecuencias de visitas más altas tanto en promedio total como en el mes actual.
Clúster 1 tiene las frecuencias más bajas.

### Distribuciones de características para los clústeres

```python
features_to_plot = [col for col in gym.columns if col not in ['cluster_km', 'Churn']]

kmean_preds = km.predict(X_test_scaled)
kmean_preds_df = pd.DataFrame(X_test_scaled, columns=features_to_plot).copy()
kmean_preds_df["cluster"] = kmean_preds
kmean_preds_df["Churn"] = y_test.reset_index(drop=True)
kmean_preds_df["cluster_from_random_forest"] = pd.qcut(rf_preds, q=5, duplicates="drop")

# Visualización de distribuciones por clúster
for col in features_to_plot:
    plt.figure(figsize=(10, 6))
    for i in range(5):
        plt.hist(kmean_preds_df[kmean_preds_df["cluster"] == i][col], alpha=0.5, label=f'group-{i}')
    plt.title(f'Distribución de {col} por Clúster')
    plt.legend()
    plt.show()
    
```

Visualizamos con mas claridad a través de estos gráficos como se distribuyen las observaciones para una característica específica en cada grupo.
1.  'Near_Location: claramente se destacan los usuarios del grupo 1 con ubicación cercana al gimnasio, en diferencia el grupo 3 que no lo están.
2. Partner y promo_friends: EL grupo 2 para ambas caracteristicas contiene los usuarios que han sido beneficiados con algún descuento a diferencia del grupo 1 y 3 que no lo son.
3. Usuarios con periodos altos hasta el fin del contrato como lo podemos ver en el grupo 2 a diferencia del grupo 1


### Tasa de cancelación para cada Clusters

```python
# Calcular la tasa de cancelación para cada clúster
cancelation_rates = kmean_preds_df.groupby("cluster")["Churn"].mean()
cancelation_rates
```

Hay diferencias significativas entre los clústeres en términos de tasas de cancelación, veamos en detalle:

1. Clúster 0 (Tasa de Cancelación: 6.92%): Este grupo tiene una tasa de cancelación relativamente baja, lo que sugiere que los usuarios en este clúster son leales y menos propensos a cancelar.

2. Clúster 1 (Tasa de Cancelación: 49.45%): Este grupo tiene la tasa de cancelación más alta, lo que indica que los usuarios en este clúster son muy propensos a cancelar.
